### Summary:
If you can see the outlier in both 'funding_total_usd' and 'funding_rounds'. So, let's find them and drop it.

    1. Find the IQR (Interquartile Range)
    2. Find the upper and lower limit
    3. Find outliers
    4. Drop them
    5. Compare the plots after trimming 


#### 2.b.1. Find the IQR

```python
company.funding_rounds.describe()
```

```python
# For funding_total_usd
#Type your code here!
q1_funding_usd = company.funding_total_usd.quantile(0.25)
q3_funding_usd = company.funding_total_usd.quantile(0.75)

IQR_funding_usd = q3_funding_usd-q1_funding_usd
print('The IQR of Funding_Total_usd is',IQR_funding_usd)

# For funding_rounds
#Type your code here!
q1_funding_rounds = company.funding_rounds.quantile(0.25)
q3_funding_rounds = company.funding_rounds.quantile(0.75)

IQR_funding_rounds = q3_funding_rounds-q1_funding_rounds
print('The IQR of Funding_Round is',IQR_funding_rounds)
```

#### 2.b.1. Find the Upper and Lower limit

```python
# For funding_total_usd
#Type your code here!

low_lim_usd = q1_funding_usd - 1.5 * IQR_funding_usd
up_lim_usd = q3_funding_usd + 1.5 * IQR_funding_usd
print('low_limit_usd is', low_lim_usd)
print('up_limit_usd is', up_lim_usd)

# For funding_rounds
#Type your code here!
low_lim_rounds = q1_funding_rounds - 1.5 * IQR_funding_rounds
up_lim_rounds = q3_funding_rounds + 1.5 * IQR_funding_rounds
print('low_limit_rounds is', low_lim_rounds)
print('up_limit_rounds is', up_lim_rounds)
```

#### 2.b.1.  Find outliers

```python
# For funding_total_usd
#Type your code here!
outlier_usd =[]
for x in company.funding_total_usd:
    if ((x> up_lim_usd) or (x<low_lim_usd)):
         outlier_usd.append(x)
print('outlier in the usd is', outlier_usd)

# For funding_rounds
#Type your code here!
outlier_rounds =[]
for x in company.funding_rounds:
    if ((x> up_lim_rounds) or (x<low_lim_rounds)):
         outlier_rounds.append(x)
print('outlier in the rounds is', outlier_rounds)

print('outlier_usd length: ',len(outlier_usd))
print('outlier_rounds length: ',len(outlier_rounds))
```

#### 2.b.1. Drop the outliers

```python
# For funding_total_usd
#Type your code here!
company.drop(company[company.funding_total_usd.isin(outlier_usd) ].index,inplace=True)

# For funding_rounds
#Type your code here!
company.drop(company[company.funding_rounds.isin(outlier_rounds) ].index,inplace=True)
```

```python
# Lets keep the category as same of it falls under above 10 otherwise let's replace it with other.

def get_category(actual_category):
  category_list = ['software','web','ecommerce','other','mobile','advertising','games_video','consulting','enterprise','biotech','public_relations','hardware','network_hosting','education','search']

  if actual_category in category_list:
    return actual_category
  else:
    return 'other'
```
